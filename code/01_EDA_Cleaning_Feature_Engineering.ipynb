### Feature Engineering

#### Duration Formatting

```python
# function to change time to decimal format in hours
def duration_to_decimal(duration):
    hours = float(duration[:2])
    minutes = float(duration[3:5])
    seconds = float(duration[6:8])
    return hours + (minutes / 60) + (seconds / 60 / 60)
```

```python
# change duration column to float in hours
combined_data['duration'] = [duration_to_decimal(i) for i in combined_data['duration']]
```

#### Chats By Duration

```python
# function to calculate chats/duration
def chats_per_hour(num_chats, duration):
    return num_chats / duration 
```

```python
combined_data['chats_per_hour'] = [chats_per_hour(num_chats, duration) for num_chats, duration in zip(combined_data['num_chats'], combined_data['duration'])]

combined_data['chats_per_hour'].head()
```

#### Sony Flag

```python
# create function to check for sony consoles
def check_Sony(clean_console_list):
    if pd.isnull(clean_console_list):
        return 0
    elif 'Playstation' in clean_console_list:
        return 1
    elif 'PS2' in clean_console_list:
        return 1
    elif 'PS3' in clean_console_list:
        return 1
    elif 'PS4' in clean_console_list:
        return 1
    elif 'PS5' in clean_console_list:
        return 1
    elif 'PSP' in clean_console_list:
        return 1
    elif 'Vita' in clean_console_list:
        return 1
    else:
        return 0
```

```python
# create is_Sony column using function above
combined_data['is_Sony'] = [check_Sony(i) for i in combined_data['initial_consoles_clean']]

print(combined_data['is_Sony'].count())
print(combined_data['is_Sony'].value_counts())
```

```python
# using numerical features
numerical_features = combined_data.select_dtypes(include = ['number'])

# calc correlations
correlations = numerical_features.drop(columns = ['id',
                                                        'was_on_PC',
                                                        'initial_release_year',
                                                        'total_donations',
                                                        'max_viewers',
                                                        'num_chats']).corr()

# pull only target variable correlations
dpm_correlations = correlations['donations_per_min'].drop('donations_per_min').sort_values(ascending=False)

# plot correlations
plt.figure(figsize = (16, 8))
sns.barplot(x = dpm_correlations.index, y = dpm_correlations.values, color = 'blue')
plt.title('Correlation of Select Features with Donations Per Minute')
plt.xlabel('Select Features')
plt.xticks(rotation = 45)
plt.ylabel('Correlation Coef')
plt.savefig('../graphics/dpm_select_correlations.jpeg')
plt.show()
```
